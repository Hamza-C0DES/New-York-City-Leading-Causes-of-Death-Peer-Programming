```python
# AI WAS UTILIZED IN THE PRODUCTION OF THIS CELL.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Clean the 'Deaths' column
# Coerce errors to convert '.' or 'F' to NaN, then fill them with 0
v_raw_data['Deaths'] = pd.to_numeric(v_raw_data['Deaths'], errors='coerce').fillna(0)

# 2. Standardize 'Leading Cause' text 
# (This prevents 'Heart Disease' and 'HEART DISEASE' from being treated as different causes)
v_raw_data['Leading Cause'] = v_raw_data['Leading Cause'].str.strip().str.upper()

# 3. Aggregate the deaths by Leading Cause
grouped_data = (
    v_raw_data.groupby('Leading Cause')['Deaths']
    .sum()
    .reset_index()
    .sort_values(by='Deaths', ascending=False)
)

# Filter for the top 10 to keep the chart clean and highly readable
top_10_causes = grouped_data.head(10)

# 4. Plotting
plt.figure(figsize=(12, 8))
sns.set_style('whitegrid')
sns.set_context("talk", font_scale=0.8) # Adjust scale so text labels fit nicely

# Create horizontal bar plot
ax = sns.barplot(
    x='Deaths', 
    y='Leading Cause', 
    data=top_10_causes, 
    palette='viridis'
)

# Set labels and title
ax.set_title("Top 10 Leading Causes of Death in NYC", fontsize=16, pad=20, weight='bold')
ax.set_xlabel("Total Deaths", fontsize=12, labelpad=10)
ax.set_ylabel("Leading Cause", fontsize=12, labelpad=10)

# Add values to the end of each bar for clarity
# for p in ax.patches:
#     width = p.get_width()
#     if width > 0:
#         ax.annotate(
#             f"{int(width):,}", # Format as a neat integer with commas
#             (width, p.get_y() + p.get_height() / 2), 
#             xytext=(10, 0),                           # Give it a 10px cushion
#             textcoords='offset points',
#             va='center',
#             fontsize=10,
#             weight='semibold'
#         )

# Adjust layout so long y-axis cause names don't get cut off
plt.tight_layout()
plt.show()
```
